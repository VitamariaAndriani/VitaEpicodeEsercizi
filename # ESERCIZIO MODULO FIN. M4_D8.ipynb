```python
# PUNTO NUM. 3

def confronta_statistiche_covid(df, continente1, continente2):
    
    # Filtriamo i dati per i due continenti specificati
    df_continente1 = df[df['continent'] == continente1]
    df_continente2 = df[df['continent'] == continente2]
    
    # Calcoliamo i descrittori statistici per i due continenti
    stats_continente1 = df_continente1['total_cases'].describe()
    stats_continente2 = df_continente2['total_cases'].describe()

    # Calcoliamo la percentuale rispetto al totale mondiale (anche considerando le locazioni senza continente)
    total_cases_mondo = df['total_cases'].sum()
    percentuale_continente1 = (stats_continente1['mean'] / total_cases_mondo) * 100
    percentuale_continente2 = (stats_continente2['mean'] / total_cases_mondo) * 100

    return {
        continente1: {
            'Minimo': stats_continente1['min'],
            'Massimo': stats_continente1['max'],
            'Media': stats_continente1['mean'],
            'Percentuale rispetto al mondo': percentuale_continente1},
        continente2: {
            'Minimo': stats_continente2['min'],
            'Massimo': stats_continente2['max'],
            'Media': stats_continente2['mean'],
            'Percentuale rispetto al mondo': percentuale_continente2}}

# Esempio di utilizzo della funzione
statistiche = confronta_statistiche_covid(owid_covid_data, 'Europe', 'South America')

# Stampiamo le statistiche
for continente, valori in statistiche.items():
    print(f"Statistiche per {continente}:")
    for k, v in valori.items():
        print(f"{k}: {v}")
    print()

# Questa funzione, confronta_statistiche_covid, prende in input il DataFrame df e due nomi di continenti. 
# Filtra il DataFrame per ottenere i dati relativi ai due continenti specificati e quindi calcola i descrittori statistici 
# (minimo, massimo, media). Successivamente, calcola la percentuale rispetto al totale mondiale, 
# considerando anche le locazioni senza indicazione di continente.
# L'esempio di utilizzo alla fine mostra come chiamare la funzione e stampare le statistiche per i due continenti specificati.
```
