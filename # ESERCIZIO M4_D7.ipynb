```python
# ESERCIZIO 2
""" Quante recensioni parlano di app?
  • Quante parlano di giochi?
  • In generale quante recensioni raccomandano il prodotto?
  • E quante non lo raccomandano?
  • Stampate qualche esempio dell'una e dell'altra.
  • Le recensioni che menzionano il Kindle sono perlopiù positive o negative?"""

import pandas as pd
db=pd.read_csv(r"C:/Users/VITA MARIA\Desktop/amazon.csv", sep=",")
amazon_db = pd.DataFrame(db)
amazon_db

# Quante recensioni parlano di app?
colonna = 'reviewText'
ricerca_valore = 'app'
filtro = amazon_db[colonna].str.contains(ricerca_valore, case=False)
risultato = amazon_db[filtro].sum()
risultato

# Quante parlano di giochi?
colonna = 'reviewText'
ricerca_valore = 'game'
filtro = amazon_db[colonna].str.contains(ricerca_valore, case=False)
risultato = amazon_db[filtro].sum()
risultato

# In generale quante recensioni raccomandano il prodotto?
colonna = 'Positive'
conteggio_valori = amazon_db[colonna].value_counts()
conteggio_valori

# E quante non lo raccomandano?
filtro_recommend = amazon_db['reviewText'].str.contains('recommend', case=False)
filtro_positive = amazon_db['Positive'] == True
risultato = amazon_db[filtro_recommend & filtro_positive ].value_counts()
risultato
risultato.sample(10)

# Stampate qualche esempio dell'una e dell'altra.
filtro_positive = amazon_db['Positive'] == False
risultato = amazon_db[filtro_positive].value_counts()
risultato
risultato.sample(10)

# Le recensioni che menzionano il Kindle sono perlopiù positive o negative?
filtro_all = amazon_db['reviewText'].str.contains('kindle', case=False)
risultato_overall = amazon_db[filtro_all]
print(risultato_overall.shape[0])

filtro = amazon_db['reviewText'].str.contains('kindle', case=False)
filtro_positive = amazon_db['Positive'] == True
risultato = amazon_db[filtro & filtro_positive]
print(risultato. shape[0])

print(int(risultato_overall.shape[0]) - int(risultato.shape[0]))

if risultato.shape[0] < int(risultato_overall.shape[0]):
    print("Le recensioni sono perlopiù positive.")

else:

    print("Le recensioni sono perlopiù negative.")
```

```python
# ESERCIZIO 4
"""Il dataset insurance.csv contiene dati rispetto a caratteristiche e abitudini 
delle persone, e della zona in cui vivono, rispetto ai costi individuali per le cure 
mediche come premio per le assicurazioni sulla salute.
- Visualizziamone le dimensioni, un'anteprima e osserviamo i nomi di colonna;
- Quali sono le medie di charges rispetto a region? 
- Ci sono differenze significative?
- E rispetto a smoker? E a sex?
- Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di 
  charges rispetto ai diversi quartili dei valori di bmi?"""

import pandas as pd
db=pd.read_csv(r"C:/Users/VITA MARIA/Desktop/insurance.csv", sep=",")
insurance= pd.DataFrame(db)
insurance

# Visualizziamone le dimensioni, un'anteprima e osserviamo i nomi di colonna
insurance.shape
insurance.sample(7)
insurance.columns

# Quali sono le medie di charges rispetto a region
medie_per_region = db.groupby('region')['charges'].mean()
medie_per_region = db

# Verificare se ci sono differenze significative
differenze_region = db.groupby('region')['charges'].std()

# Calcolare le medie di charges rispetto a smoker e sex
medie_per_smoker = db.groupby('smoker')['charges'].mean()
medie_per_sex = db.groupby('sex')['charges'].mean()

# Calcolare i descrittori statistici di bmi
desc_bmi = db['bmi'].describe()

# Calcolare minimo, media e massimo di charges rispetto ai diversi quartili di bmi
quartili_bmi = pd.qcut(db['bmi'], 4) # suddivide il range in 4 parti uguali 
statistiche_per_quartile = db.groupby(quartili_bmi)['charges'].agg(['min', 'mean', 'max'])

# Stampare i risultati
print("\nMedie di charges rispetto a region:\n", medie_per_region)
print("\nDifferenze significative per region:\n", differenze_region)
print("\nMedie di charges rispetto a smoker:\n", medie_per_smoker)
print("\nMedie di charges rispetto a sex:\n", medie_per_sex)
print("\nDescrittori statistici di bmi:\n", desc_bmi)
print("\nStatistica di charges rispetto ai quartili di bmi:\n", statistiche_per_quartile)

```
